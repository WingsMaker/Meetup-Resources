Function to obtain list of titles written in a file.
where titles are seperated by line.

```python
def Obtain_Title_list(Source_file_name):
    f=open(Source_file_name,"r")
    title_list=f.read().splitlines()
    f.close()
    return [t for t in title_list if t.strip()]
```

```python
def write_to_csv(file_Name,S_S):
    csv_file=open(file_Name,"w")
    w=csv.writer(csv_file,dialect='excel')
    w.writerows(S_S)
    csv_file.close()
```

Indeed Job search Using INDEED API
This function searches specified term in Indeed Using Indeed API and returns the result as list of dictionaries with snippet, title, URL, Published date, etc.

```python
def getJob(**kwargs):
    '''getJob(q=job title,start=start from,format=json|xml,...)'''
    args = dict(publisher=1288266145526105, v=2,format="json", highlight=0, filter=0, limit=10)
    args.update(**kwargs)
    url = 'http://api.indeed.com/ads/apisearch?' + urlencode(args)
    jobs= json.load(urlopen(url,proxies=proxies))
    return jobs['results']
```

function returns content of page/URL specified

```python
def getpage(url):
    return urlopen(url,proxies=proxies).read()
```

This function takes a indeed page as HTML and Extracts only Summary part using BeautifulSoup module and returns Extracted summary.

```python
def getSummary(content):
    Soup=BeautifulSoup(content)
    summary=Soup.find(name="span",attrs={"class":"summary"})
    return summary.getText(separator='. ')
```

```python
Stop_words={}
less_than_3_char_skills=["bi","c","c#","e#","es","f#","j#","ml","r","vb","x#"]
```

```python
def uniqeSet_with_count(r,l,vocab):
	doc={}
	for i in l:
		W=r[2].sub("",(i.strip("-\t\n\r ") if i else ""))
	 	if len(W)>=3 or W in less_than_3_char_skills:
			if W in doc:
				doc[W]+=1
			else:
				if not W in vocab:
					vocab.append(W)
				doc[W]=1
	return doc
```

Function that detects whether a word is stopword or not.

```python
def is_StopWord(x):
    l=len(x)
    if l in Stop_words and x in Stop_words[l]:
        return True
    elif "-" in x:
        return all(is_StopWord(i) for i in x.split("-"))
    elif l>5:
        return any((is_StopWord(x[:i]) and is_StopWord(x[i:])) for i in range(2,l-1) if (i-1) in Stop_words and (l-i) in Stop_words )
    else:
        return False
```

Function that updates vocabulary by elemination stopwords. 

```python
def update_vocab(V):
   return tuple(W for W in  V if not is_StopWord(W.strip("-0123456789")))
```

Splits the string into vocabulary.

```python
def split(r,x):
    return [i for i in r[0].split(r[1].sub(r"\1-\2",x)) if i]
```

```python
def main(input_file="topic_list.txt",Num_topics=10,Stop_word="Stop_words.txt"):
    global Stop_words
    #r=re.compile(r'[^a-zA-Z#\+\-]+')
    r=[]
    r.append(re.compile(r'[^a-zA-Z#\+\-0-9]+|(?<![a-zA-Z])[0-9]+(?![a-zA-Z])'))
    r.append(re.compile(r"([a-zA-Z])\s*\-\s*([a-zA-Z])"))
    r.append(re.compile(r"^[0-9\-#\+]"))
    Stop_word_file=open(Stop_word,"r")
    for i in sorted(Stop_word_file.read().split(),key=lambda l:len(l)):
        length=len(i)
        if length in Stop_words:
            Stop_words[length].append(i)
        else:
            Stop_words[length]=[i]
    Stop_word_file.close()
    vocab=[] #vocab Uniqe words
    Titles=Obtain_Title_list(input_file)
    #Jobs=get_JobDetails()
    Jobs=[]
    for title in Titles:
        title = re.compile(r'^\s*|\s*$').sub('',title)
        queted_title = 'title:("' + title + '")' if '")' not in title else title
        Result=getJob(q=queted_title.encode('ascii', 'ignore'),start=0, limit=5)
        print title,len(Result)
        Jobs+=Result
    for i in Jobs:
        i['summary']=getSummary(getpage(i["url"]))
    documents= [uniqeSet_with_count(r,split(r,J["jobtitle"].lower())+split(r,J["summary"].lower()),vocab) for J in Jobs]#set of all document with word and frequency of occurence
    s={i:[k for j in i.split("-") for k in (r[2].sub("",j),) if k in vocab] for i in vocab if "-" in i}
    s={i:s[i] for i in s if s[i]}
    for d in documents:
        for w in list(d):
            if "-" in w:
                for nw in [j for i in w.split("-") for j in (r[2].sub("",i),) if j in vocab]:
                    if nw in d:
                        d[nw]+=1
                    else:
                        d[nw]=1  
        for i in s:
            if all([(j in d) for j in s[i]]):
                if i in d:
                    d[i]+=1
                else:
                    d[i]=1                 
    vocab=update_vocab(vocab)
    n_documents= len(documents)#number of documents
    vocab_size=len(vocab)#number of uniqe Words
    model = lda.LDA(n_topics=Num_topics, n_iter=1000, random_state=1)
    X=np.ndarray((n_documents,vocab_size), 
            buffer=np.array([(documents[i][vocab[j]] if vocab[j] in documents[i] else 0) for i in range(n_documents) for j in range(vocab_size)]),
            offset=0,dtype=int)
    model.fit(X)
    topic_word = model.topic_word_
    n_top_words = 25
    topic_top_words=[["Topic ID"]+["Vocab {}".format(i+1) for i in range(n_top_words)]]
    for i, topic_dist in enumerate(topic_word):
        topic_words = np.array(vocab)[np.argsort(topic_dist)][:-(n_top_words+1):-1]
        topic_top_words.append(["Topic {:2}".format(i+1)]+[t.encode('utf-8',"ignore") for t in topic_words])
    write_to_csv("Topics_top_%s_words(indeed).csv" % n_top_words,topic_top_words)
    doc_topic = model.doc_topic_
    Topic_grouping=[]
    for i in range(n_documents):
        Topic_grouping.append("Topic {:2}".format(doc_topic[i].argmax()+1))
    heading=[["Topic ID","JOB Title","URL","Job Description"]]
    Topic_grouping=sorted([[Topic_grouping[i],Jobs[i]["jobtitle"].encode('ascii',"ignore"),Jobs[i]["url"],Jobs[i]['summary'].encode('ascii',"ignore")] for i in range(n_documents)],key=lambda l:l[0])
    write_to_csv("Topics_Job_Description(indeed).csv",heading+Topic_grouping)
    return topic_word, doc_topic,Jobs,vocab
```

```python
if __name__ == '__main__':
   topics,docs,Jobs,vocab= main(input_file="Topics",Num_topics=5,Stop_word="Stop_words.txt")    
```
